```python
import matplotlib.pyplot as plt

df["Major_Category"].value_counts().plot(kind="bar")

plt.title("Students by Major Category")
plt.xlabel("Major")
plt.ylabel("Number of Students")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.hist(df["Pre_Semester_GPA"], bins=10)

plt.title("Distribution of Pre Semester GPA")
plt.xlabel("GPA")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(df["Post_Semester_GPA"], bins=10)

plt.title("Distribution of Post Semester GPA")
plt.xlabel("GPA")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(df["Weekly_GenAI_Hours"], df["Post_Semester_GPA"])

plt.title("Weekly GenAI Hours vs Post GPA")
plt.xlabel("Weekly GenAI Hours")
plt.ylabel("Post Semester GPA")
plt.show()
```

```python
plt.scatter(df["Traditional_Study_Hours"], df["Post_Semester_GPA"])

plt.title("Study Hours vs GPA")
plt.xlabel("Traditional Study Hours")
plt.ylabel("Post Semester GPA")
plt.show()
```
